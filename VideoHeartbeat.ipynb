```python
# pixel manipulation of images

def IncreaseColourValue(colourValue):
    colourValue *= 2
    if (colourValue > 255):
        colourValue = 255
    return colourValue

#---------------------------------------------------------------------------------------------


def AmplifyColour(colourValue):
    maxColourCoord = 0
    r = colourValue[0]
    g = colourValue[1]
    b = colourValue[2]
    if (r >= g):
        if (r >= b):
            # mostly red
            r = IncreaseColourValue(r)
        else:
            # mostly blue
            b = IncreaseColourValue(b)
    else:
        # mostly green
        g = IncreaseColourValue(g)
        
    return [r, g, b] 

#----------------------------------------------------------------------------------------------

def ProcessImages():
    # images = glob.glob("Development/Projects/Python/JupyterNotebooks/VideoHeartbeat/TestImages/*.jpg")
    directory_in_str = "C:/Users/Frank_Work/Development/Projects/Python/JupyterNotebooks/VideoHeartbeat/TestImages/"
    directory = os.fsencode(directory_in_str)
    
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.endswith(".jpg"): 
            directoryName = directory.decode('utf-8')
            filePath = os.path.join(directoryName, filename)
            img = Image.open(filePath)
            pix = img.load()
            for i in range(0,img.size[0]):
                for j in range(0,img.size[1]):
                    pixCol = pix[i,j]
                    pix[i,j] = AmplifyColour(pix[i,j])
            imgName = filename.split('.')[0]
            img.save('./ProcessedImages/' + imgName + 'EDITED.jpg')


```

```python
# getting frames of video
import cv2

videoDir = "C:/Users/Frank_Work/Development/Projects/Python/JupyterNotebooks/VideoHeartbeat/TestVideos/"
videoOutputDir = "C:/Users/Frank_Work/Development/Projects/Python/JupyterNotebooks/VideoHeartbeat/ProcessedVideos/"
videoName = 'BHouseBathroom - Trim.mp4'
# Open the input video
cap = cv2.VideoCapture(videoDir + videoName)

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))   # float `width`
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # float `height`
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can choose other codecs

print(total_frames)

# Define the codec and create VideoWriter object
out = cv2.VideoWriter(videoOutputDir + videoName, fourcc, fps, (width, height))

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Example manipulation: invert colors
    # edited_frame = cv2.bitwise_not(frame)
        # Example manipulation: Draw a red square in the top-left corner
    for y in range(height):
        for x in range(width):
            frame[y, x] = AmplifyColour(frame[y, x])  # BGR for red
    edited_frame = frame
    # Write the edited frame to the output video
    out.write(edited_frame)
    print(frame_count)
    frame_count+=1

# Release everything when done
cap.release()
out.release()
cv2.destroyAllWindows()
```
